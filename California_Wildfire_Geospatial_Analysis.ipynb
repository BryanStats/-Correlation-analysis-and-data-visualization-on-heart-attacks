```python
import folium
import pandas as pd
```

```python
fire_data = {'Fire_Name': ['Camp Fire', 'Dixie Fire', 'Thomas Fire'],
             'Lat': [39.8134, 40.0333, 34.4253],
             'Lon': [-121.4347, -121.2225, -119.0645],
             'Acres': [153336, 963309, 281893]}
df = pd.DataFrame(fire_data)
```

```python
ca_fire_map = folium.Map(location=[37.5, -119.5], zoom_start=6, tiles='CartoDB dark_matter')
```

```python
for idx, row in df.iterrows():
    folium.CircleMarker(
        location=[row['Lat'], row['Lon']],
        radius=row['Acres'] / 40000,
        popup=f"{row['Fire_Name']}: {row['Acres']:,} acres",
        color='#ff4500',
        fill=True,
        fill_color='#ff4500',
        fill_opacity=0.5  # <--- This makes overlapping markers visible!
    ).add_to(ca_fire_map)
```

```python
ca_fire_map.save('CA_Wildfire_Risk_Map.html')
```
